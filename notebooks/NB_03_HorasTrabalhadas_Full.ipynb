```python
from pyspark.sql.types import *

# Define o schema do dataframe
csv_schema = StructType([ \
    StructField("Matricula", IntegerType(), True), \
    StructField("Data", DateType(), True), \
    StructField("HorasNormais", DoubleType(), True), \
    StructField("HorasTrabalhadas", DoubleType(), True)
  ])

# Lê o arquivo
df = spark.read.format("csv") \
    .option("header", "true") \
    .option("delimiter", ";") \
    .option("dateFormat", "dd/MM/yyyy") \
    .schema(csv_schema) \
    .load("Files/horas_trabalhadas.csv")

display(df)

# Salva nas tabelas delta
df.write.mode('overwrite').saveAsTable("staging_horas_trabalhadas")
```

```python
%%sql

-- Apaga a tabela se existir
DROP TABLE IF EXISTS tb_horas_trabalhadas;

-- Cria a tabela a partir incluindo a 'MatriculaSk' da tabela 'tb_funcionarios'
CREATE TABLE tb_horas_trabalhadas AS
SELECT 
    h.Matricula, 
    h.Data, 
    h.HorasNormais, 
    h.HorasTrabalhadas,
    f.MatriculaSk
FROM 
    staging_horas_trabalhadas h

LEFT JOIN tb_funcionarios f
    ON h.Matricula = f.Matricula
    AND h.Data BETWEEN f.DataVigenciaInicial AND f.DataVigenciaFinal;

SELECT * FROM tb_horas_trabalhadas;

-- Apaga a tabela temporária
DROP TABLE IF EXISTS staging_horas_trabalhadas;
```
